```python
#import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head(10) # read first 10 rows
```

```python
df.info() # get summary
```

```python
df.describe() # check statistics
```

```python
df.isnull().sum() #find empty values
```

```python
df.duplicated().sum() #find duplicate
```

Awesome, there is no null value and also no duplicates value.:::bold text

```python
df['Total'] = df['writing score'] + df['reading score'] + df['math score']
df['Average'] = df['Total']/3
df  # add new column
```

```python
df.columns # see all columns
```

```python
df = df.rename(columns={'gender':'Gender', 'race/ethnicity': 'Race/Ethnicity', 'parental level of education': 'Parental Level of Education', 'lunch':'Lunch', 'test preparation course':'Test Preparation Course', 'math score':'Math Score', 'reading score': 'Reading Score', 'writing score': 'Writing Score'})
# rename of columns
```

```python
df.groupby('Parental Level of Education').max(numeric_only=True) # used groupby
```

```python
df.groupby('Parental Level of Education').min(numeric_only=True)
```

```python
df.sort_values(by=['Race/Ethnicity', 'Gender'], inplace=True)
df # used sort_values
```

```python
race_counts = df['Race/Ethnicity'].value_counts()
plt.pie(race_counts, labels=race_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Race/Ethnicity')
plt.show()
```

```python
education_count = df['Parental Level of Education'].value_counts()
plt.pie(education_count, labels=education_count.index, colors=['yellow', 'orange', 'pink', 'green', 'brown', 'grey'], autopct='%1.1f%%')
plt.title('Distribution of Parental Level of Education')
plt.show()
```

```python
plt.bar(df['Gender'], df['Average'])
plt.xlabel('Gender')
plt.ylabel('Average')
plt.title('Average Score by Gender')
plt.show()
```

```python
df.hist(figsize=(10, 6))
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.title('Histogram of Scores')
plt.tight_layout()
plt.show()
```
